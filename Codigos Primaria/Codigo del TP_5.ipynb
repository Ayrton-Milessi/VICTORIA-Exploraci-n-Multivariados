```r
#Graficamos diagrama de dispersion entre las notas de Lengua y Matematica de Primer año
#Preparamos los datos:

#Filtramos los datos que corresponden a MATEMATICA
mat_primer_grado_<- mat_primer_grado %>%
  select(Documento, asignatura, Promedio)
#Filtramos los datos que corresponden a Lengua
leng_primer_grado_ <- leng_primer_grado %>%
  select(Documento, asignatura, Promedio)

# Resumir los datos de Matemáticas y Lengua
mat_primer_grado_summary <- mat_primer_grado_ %>%
  group_by(Documento) %>%
  summarize(Promedio_Mat = mean(Promedio))

leng_primer_grado_summary <- leng_primer_grado_ %>%
  group_by(Documento) %>%
  summarize(Promedio_Len = mean(Promedio))

# Combinar los resúmenes con left_join
calificaciones_combinadas_primero <- left_join(mat_primer_grado_summary, leng_primer_grado_summary, by = "Documento")

#Omitimos las filas con NA
calificaciones_combinadas_primero <- na.omit(calificaciones_combinadas_primero)

#Chequeamos
head(calificaciones_combinadas_primero)
dim(calificaciones_combinadas_primero)
```

```r
plot(calificaciones_combinadas_primero$Promedio_Mat, calificaciones_combinadas_primero$Promedio_Len, main="Grafico de dispersion",
    xlab = "Prom. Matematica 1er año", ylab = "Prom. Lengua 1er año", pch = 19, col = "blue")
abline(lm(calificaciones_combinadas_primero$Promedio_Len~calificaciones_combinadas_primero$Promedio_Mat), col ="red")
```

```r
#Se quiere estudiar si existe correlación entre las calificaciones en Matematica y las calificaciones en Educacion Física
#Preparamos los datos:

#Filtramos los datos que corresponden a MATEMATICA
calificaciones_mat <- Victoria_primaria %>%
  filter(asignatura %in% c("MATEMATICA")) %>%
  select(Documento, asignatura, Promedio)
#Filtramos los datos que corresponden a EDUCACION FISICA
calificaciones_ef <- Victoria_primaria %>%
  filter(asignatura %in% c("EDUCACION FISICA")) %>%
  select(Documento, asignatura, Promedio)

# Resumir los datos de Matemáticas y Educación Física
calificaciones_mat_summary <- calificaciones_mat %>%
  group_by(Documento) %>%
  summarize(Promedio_Mat = mean(Promedio))

calificaciones_ef_summary <- calificaciones_ef %>%
  group_by(Documento) %>%
  summarize(Promedio_EF = mean(Promedio))

# Combinar los resúmenes con left_join
calificaciones_combinadas <- left_join(calificaciones_mat_summary, calificaciones_ef_summary, by = "Documento")

#Omitimos las filas con NA
calificaciones_combinadas <- na.omit(calificaciones_combinadas)

#Chequeamos
head(calificaciones_combinadas)
dim(calificaciones_combinadas)
```

```r
plot(calificaciones_combinadas$Promedio_EF, calificaciones_combinadas$Promedio_Mat, main="Grafico de dispersion",
     xlab = "Promedio EF", ylab = "Promedio Mat.", pch = 19, col = "blue")
abline(lm(calificaciones_combinadas$Promedio_Mat~calificaciones_combinadas$Promedio_EF), col = "red")

shapiro.test(calificaciones_combinadas$Promedio_EF)
shapiro.test(calificaciones_combinadas$Promedio_Mat)

correlacion_spearman <- cor(calificaciones_combinadas$Promedio_EF,calificaciones_combinadas$Promedio_Mat, method = "spearman")
print(correlacion_spearman)
print("Concluímos: Hay una correlación fuerte (0.288119)")
```
