Hay materias que están relacionadas conceptualmente o temáticamente, como Matemáticas y Física, podría ser interesante medir su correlación para comprender cómo se complementan o se refuerzan mutuamente.

```r
#Se quiere estudiar si existe correlación entre las calificaciones en Matematica y las calificaciones Física
#Preparamos los datos:

#Filtramos los datos que corresponden a MATEMATICA
calificaciones_mat <- secundaria_VICTORIA %>%
  filter(asignatura %in% c("MATEMATICA")) %>%
  select(Documento, asignatura, Promedio)
#Filtramos los datos que corresponden a EDUCACION FISICA
calificaciones_fisica <- secundaria_VICTORIA %>%
  filter(asignatura %in% c("FISICA")) %>%
  select(Documento, asignatura, Promedio)

# Resumir los datos de Matemáticas y Educación Física
calificaciones_mat_summary <- calificaciones_mat %>%
  group_by(Documento) %>%
  summarize(Promedio_Mat = mean(Promedio))

calificaciones_fis_summary <- calificaciones_fisica %>%
  group_by(Documento) %>%
  summarize(Promedio_Fis = mean(Promedio))

# Combinar los resúmenes con left_join
calificaciones_mat_fis <- left_join(calificaciones_mat_summary, calificaciones_fis_summary, by = "Documento")

#Omitimos las filas con NA
calificaciones_mat_fis <- na.omit(calificaciones_mat_fis)

#Chequeamos
head(calificaciones_mat_fis)
dim(calificaciones_mat_fis)
```

```r
# Crear un gráfico de dispersión
#Para ver si tiene sentido lo que voy a hacer podemos graficar para tener una nocion previa
plot(calificaciones_mat_fis$Promedio_Fis, calificaciones_mat_fis$Promedio_Mat, main = "Gráfico de dispersión: Promedio Fisica vs Promedio Matemática",
       xlab = "Promedio EF", ylab = "Promedio Mat.", pch = 19, col = "blue")
abline(lm(calificaciones_mat_fis$Promedio_Mat ~ calificaciones_mat_fis$Promedio_Fis), col = "red")


# Prueba de normalidad (por ejemplo, prueba de Shapiro-Wilk)
shapiro_test_promedio_Fis<- shapiro.test(calificaciones_mat_fis$Promedio_Fis)
shapiro_test_promedio_matematica <- shapiro.test(calificaciones_mat_fis$Promedio_Mat)

# Impresión de los resultados de la prueba de normalidad
print("Prueba de normalidad para promedio Fisica:")
print(shapiro_test_promedio_Fis)
print("Prueba de normalidad para promedio Matemática:")
print(shapiro_test_promedio_matematica)

#me da muy chico, rechazo Ho) es una distribucion normal
print("Los datos no pasaron la prueba de normalidad.No se puede calcular la correlación de Pearson. Calculamos la Correlación de Spearman")

correlacion_spearman <- cor(calificaciones_mat_fis$Promedio_Fis, calificaciones_mat_fis$Promedio_Mat, method = "spearman")

# Impresión del resultado
print(correlacion_spearman)
print("Concluímos Hay una correlación moderada positiva entre las calificaciones obtenidas en Fisica y las obtenidas en Matematica")
```
